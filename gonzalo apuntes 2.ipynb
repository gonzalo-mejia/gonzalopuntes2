```python
from numpy import linspace, sin, cos
```

```python
x=linspace(0, 10, 1000)
y=sin(x)

plot(x,y)
show()
```

```python
x=linspace(0, 10, 1000)
y=cos(x)

plot(x,y)
show()
```

```python
from pylab import ylim, xlim, grid
```

```python
x=linspace(0, 10, 1000)
y=sin(x)

plot(x,y)
ylim(-1.1, 1.1)
xlim(0., 10)
grid("on")
show()
```

```python
from pylab import xlabel, ylabel
```

```python
plot(x,y)
ylim(-1.1, 1.1)
xlim(0., 10)
grid("on")
xlabel("eje x")
ylabel("eje y")
show()
```
